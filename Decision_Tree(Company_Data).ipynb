Labelling Sales values less than sales.mean value as **"Low"** and greater than sales.mean values as "High"

```python
sales_val = []
for value in df["Sales"]:
    if value>=7.49:
        sales_val.append("High")
    else:
        sales_val.append("low")
df["Sales_val"]    = sales_val
```

## **Data Transformation**

```python
import pandas as pd
import warnings
pd.options.mode.chained_assignment = None
warnings.simplefilter(action='ignore', category=DeprecationWarning)
import warnings
with warnings.catch_warnings():
    # Setting values in-place is fine, ignore the warning in Pandas >= 1.5.0
    # This can be removed, if Pandas 1.5.0 does not need to be supported any longer.
    # See also: https://stackoverflow.com/q/74057367/859591
    warnings.filterwarnings(
        "ignore",
        category=FutureWarning,
        message=(
            ".*will attempt to set the values inplace instead of always setting a new array. "
            "To retain the old behavior, use either.*"
        ),
    )
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cat = df[df.columns[[6,9,10]]]
df_cat.iloc[:,0] = le.fit_transform(df_cat.iloc[:,0])
df_cat.iloc[:,1] = le.fit_transform(df_cat.iloc[:,1])
df_cat.iloc[:,2] = le.fit_transform(df_cat.iloc[:,2])
df_cat
```

```python
df_cont = df.drop(df.columns[[0,6,9,10,11]],axis =1)
df_cont
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
ss = ss.fit_transform(df_cont)
ssx = pd.DataFrame(ss)
ssx.columns = list(df_cont)
ssx
```

```python
df1 = pd.concat([ssx,df_cat,df["Sales_val"]],axis =1)
df1
```

```python
x = df1.iloc[:,0:10]
y= df1["Sales_val"]
x.head()
```

## **Data Partition**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8)
```

# **Decision Tree Classifier (CART) Using "Gini" Criteria**

```python
from sklearn.tree import DecisionTreeClassifier
dct = DecisionTreeClassifier(criterion = "gini",max_depth =5)
dct.fit(x_train,y_train)
y_pred_train = dct.predict(x_train)
print(y_pred_train)
y_pred_test = dct.predict(x_test)
print(y_pred_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_train,y_pred_train)
print("The training accuracy score is:",acc.round(3))
acc = accuracy_score(y_test,y_pred_test)
print("The testing accuracy score is:",acc.round(3))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
import graphviz
data = tree.export_graphviz(dct,filled = True,rounded = True,special_characters = True,feature_names= ["CompPrice","Income","Advertising","Population","Price","Age","Education","ShelveLoc","Urban","US"], class_names= ['low', 'High'])

graph = graphviz.Source(data)
graph
```

The lowest gini value indicates the most important variable for classifier.

# Decision Tree Classifier (CART) Using "Entropy" Criteria

```python
from sklearn.tree import DecisionTreeClassifier
dct = DecisionTreeClassifier(criterion = "entropy",max_depth =5)
dct.fit(x_train,y_train)
y_pred_train = dct.predict(x_train)
print(y_pred_train)
y_pred_test = dct.predict(x_test)
print(y_pred_test)
```

```python

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_train,y_pred_train)
print("The training accuracy score is:",acc.round(3))
acc = accuracy_score(y_test,y_pred_test)
print("The testing accuracy score is:",acc.round(3))
```

```python
from sklearn import tree
import graphviz
data = tree.export_graphviz(dct,filled = True,rounded = True,special_characters = True,feature_names= ["CompPrice","Income","Advertising","Population","Price","Age","Education","ShelveLoc","Urban","US"], class_names= ['low', 'High'])
graph = graphviz.Source(data)
graph
```

The Entropy "0" indicates the purity of the variable
